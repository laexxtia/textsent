```python
from sklearn.feature_selection import SelectKBest, f_classif

# Assuming you have a target variable 'label' in your features DataFrame
X = features.drop(columns=['compound','class'])  # Adjust if you have additional columns to exclude
y = features['class']  # Adjust the target column name as necessary

# Apply feature selection
selector = SelectKBest(f_classif, k=10)
selector.fit(X, y)
top_features = X.columns[selector.get_support(indices=True)]
print("Top predictive features:", top_features)
```

<h3>Comparing Offensive & Hate Speech Bigrams

```python
from collections import Counter
import seaborn as sns
import nltk

def plot_ngrams(data, n=1):
    all_text = ' '.join(data['tweet'])
    ngrams = (pd.Series(nltk.ngrams(all_text.split(), n)).value_counts())[:10]
    ngrams.sort_values().plot.barh(color='blue', width=0.9, figsize=(12, 8))
    plt.title(f'Top {n}-grams')
    plt.show()

plot_ngrams(data[data['class'] == 0], 2)  # For bigrams in hate speech
```

```python
plot_ngrams(data[data['class'] == 1], 2)  # For unigrams in offensive speech
```
